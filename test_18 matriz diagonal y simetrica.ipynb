# **Inversa**

```python
A_inv = np.diag([1/2,1/3,1/4,1/5])
identidad = A.dot(A_inv)
print(identidad)
```

Tambien se puede calcular por:

```python
A_inv_calc = np.linalg.inv(A)
print(A_inv_calc)
```
