# **Sin solucion**


*  y= 3x+5
*  y = -1+3
*  y = 2x+1

.arange : genera un conjunto de nmeros entre un valor de inicio y uno final, pudiendo especificar un incremento entre los valores, pero, al contrario de lo que ocurre con range, el resultado aqu es un array NumPy:



tutorial_numpy_0004.jpg

```python
x = np.arange(-6, 6)

y_1 = 3*x+5
y_2 = -x+3
y_3 = 2*x+1
```

```python
plt.figure()
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

#limite de los ejes
plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color = "grey")
plt.show()

```

No se cruzan en un mismo punto las tres ya que tenemos mas ecuaciones que variables: **No tiene solucion**

# **Con solucion**

```python
x = np.arange(-6,6)

y_2 = -x+3
y_3 = 2*x+1
```

```python
plt.figure()
plt.plot(x, y_2)
plt.plot(x, y_3)

#limite de los ejes
plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color = "grey")
plt.show()
```
