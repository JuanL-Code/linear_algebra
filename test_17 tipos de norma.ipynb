* L0 = # Elementos del vector distinto de cero
* L1 = Sumatoria  hasta i del valor absolutro de los elementos del array
* L2 = Distancia euclidiana, distancia al origen
* LInfinito = mayor valor dentro de los valores absolutos de nuestro vector  

```python
import numpy as np
```

**Norma cero (L0)**

```python
vector = np.array([-50,-25,0,25,100,-300])
#Norma cero (L0), elementos distintos de cero
print (np.linalg.norm(vector, ord = 0))
```

**Norma uno (L1)**

```python
vector = np.array([1,-1,1,-1,1])
#Suma de valores absolutos
print(np.linalg.norm(vector, ord = 1))
```

**Norma dos (L2)**

```python
vector = np.array([1/2,1/2,1/2,1/2])
```

```python
print(np.linalg.norm(vector))
print(np.linalg.norm(vector, ord = 2))
```

**L2 cuadrado**


```python
# En machine learning se usa mucho la norma L2 pero elevada al cuadrado
# es decir, L2^2. La ventaja de esto es computacional.
# Podríamos reunir muchos vectores en una matriz X, calcular X.dot(x_t)
# y tendríamos allí el cuadrado de las normas de todos esos vectores.
```

```python
vector = np.array([1,2,3,4,5,6])
print(np.linalg.norm(vector, ord = 2)**2)
```

```python
print(vector.T.dot(vector))
```

```python
vector = np.array([1,2,3,-100])
print(np.linalg.norm(vector, ord = np.inf))
```
